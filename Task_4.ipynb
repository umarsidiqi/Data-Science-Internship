```python
import pandas as pd
import numpy as np

data = {
    'User': ['U1','U1','U1','U2','U2','U2','U3','U3','U3'],
    'Movie': ['M1','M2','M3','M1','M2','M3','M1','M2','M3'],
    'Rating': [5,4,0,4,5,3,2,3,5]
}

df = pd.DataFrame(data)
df

user_item = df.pivot(index='User', columns='Movie', values='Rating')
user_item

predicted_ratings = user_item.fillna(user_item.mean())
predicted_ratings

print("Predicted rating for User U1 on Movie M3:",
      predicted_ratings.loc['U1', 'M3'])
```
